# **Aula 7: Machine Learning com Scikit-learn**

**Scikit-learn - Etapa 1**

```python
# Importando da biblioteca sklearn
from sklearn.datasets import make_regression
```

```python
# Criando uma regressão (amostra de dados)
x, y = make_regression(n_samples=200, n_features=1, noise=30)
```

```python
 # Importando da biblioteca matplotlib
 import matplotlib.pyplot as plt
```

```python
# Criando um gráfico dessa regressão
plt.scatter(x, y);
```

```python
# Importando da biblioteca sklearn
from sklearn.linear_model import LinearRegression
```

```python
# Criando um modelo de Machine Learning
modelo = LinearRegression()
```

```python
# Inserindo os dados ao modelo
modelo.fit(x,y)
```

```python
# Prevendo os valores do nosso modelo (Machine Learning)
modelo.predict(x)
```

```python
# Construíndo um gráfico desse modelo
plt.scatter(x,y)
plt.plot(x, modelo.predict(x), color="red", linewidth=3)
```
