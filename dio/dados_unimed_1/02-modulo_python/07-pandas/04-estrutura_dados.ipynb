### **Listas**

```python
# Criando uma lista chamada animais
animais = [1,2,3]
animais
```

```python
# Sobrescrevendo a lista animais
animais = ["cachorro", "gato", 12345, 6.5]
```

```python
# Imprimindo o primeiro elemento da lista
animais[0]
```

```python
# Imprimindo o 4 elemento da lista
animais[3]
```

```python
# Substituindo o primeiro elemento da lista
animais[0] = "papagaio"
animais
```

```python
# Removendo gato da lista
animais.remove("gato")
animais
```

```python
# Verificando o tamanho da lista (Quantidade de elementos)
len(animais)
```

```python
# Verificando se gato existe na lista
"gato" in animais
```

```python
# Adicionando um elemento a lista
animais.append("leão")
animais
```

```python
# Inserindo uma nova informação na posição (index) determinada
animais.insert(1, "pato")
animais
```

```python
# Extendendo a lista
animais.extend(["cobra", 6])
animais
```

```python
# Verificando a quantidade de um elemento em uma lista
animais.count("cachorro")
```

```python
# Verificando o index (índice) de um elemento
animais.index("leão")
```

```python
# Exibindo a lista
animais
```

```python
# Removendo um item da lista pelo index
animais.pop(2)
animais
```

```python
# Criando outra lista
lista = [500, 30, 300, 80, 10]
lista
```

```python
# Verificando o valor máximo da lista
max(lista)
```

```python
# Verificando o valor mínimo da lista
min(lista)
```

```python
# Ordenando uma lista em ordem crescente
lista.sort()
lista
```

```python
# Ordenando uma lista em ordem contrária a atual
lista.reverse()
lista
```

### **Tuplas**

```python
# Criando uma tupla (As tuplas usam parênteses como sintaxe)
tp = ("Banana", "Maçã", 10, 50)
tp
```

```python
# Retornando o primeiro elemento da tupla
tp[0]
```

```python
# Tentando alterar um elemento da tupla (Diferente das listas as tuplas são imutáveis, o que quer dizer que não podemos alterar os seus elementos)
tp[0] = "Laranja"
```

```python
# Verificando quantos elementos determinado existe na tupla
tp.count("Maçã")
```

```python
# Fatiando a tupla
tp[0:2]
```

### **Dicionários**

```python
# Criando um dicionário (utilizamos as {}, eles trabalham com o conceito chave e valor)
dc = {"Maçã":20, "Banana":10, "Laranja": 15, "Uva": 5}
dc
```

```python
# Acessando o valor de um dicionário através da chave
dc["Maçã"]
```

```python
# Atualizando o valor da Maçã
dc["Maçã"] = 25
dc
```

```python
# Retornando todas as chaves do dicionário
dc.keys()
```

```python
# Retornando os valores do dicionário
dc.values()
```

```python
# Verificando se já existe uma chave no dicionário e caso não exista, inserir
dc.setdefault("Limão", 22)
dc
```
