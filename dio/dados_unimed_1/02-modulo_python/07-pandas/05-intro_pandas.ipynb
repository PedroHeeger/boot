```python
# Importando a bilioteca pandas
import pandas as pd
```

```python
# Criando um Dataframe
df = pd.read_csv("/content/drive/MyDrive/4 PROJ/atual/boot/dio/dados_unimed_1/Cusro_Python_Pandas_Digital_Innovation-master/datasets/Gapminder.csv", error_bad_lines=False, sep=";")
```

```python
# Renomeando o nome das colunas
df = df.rename(columns={"country":"País", "continent": "Continente", "year":"Ano", "lifeExp": "Expectativa de vida", "pop": "Pop Total", "gdpPercap": "PIB"})
df.head()
```

```python
# Filtrando em um Dataframe
Oceania = df.loc[df["Continente"] == "Oceania"]
Oceania.head()
```

```python
# Agrupando um Dataframe
df.groupby("Continente")["País"].nunique()
```
